## Feature Transformation

this feature transformation is for research question 2

```python
last_10_years_esports = esport_earnings_df[esport_earnings_df["Releaseyear"]>=2011]
last_10_years_esports = last_10_years_esports[last_10_years_esports["TotalMoney"] != 0]
last_10_years_esports = last_10_years_esports[last_10_years_esports["PlayerNo"] != 0]
last_10_years_esports
```

```python
last_10_years_esports = last_10_years_esports.copy()
```

```python
last_10_years_esports["YearsAfterRelease"] = 2021 - last_10_years_esports["Releaseyear"]
```

```python
column_names = ['GameName', 'Releaseyear', 'YearsAfterRelease', 'Genre', 'TotalMoney', 'PlayerNo','TournamentNo']

# Reorganize the columns
last_10_years_esports = last_10_years_esports.reindex(columns=column_names)
```

```python
last_10_years_esports
```

```python
esports_sorted = last_10_years_esports.sort_values('TotalMoney', ascending=False).copy()
```

```python
esports_sorted.head(15)
```

```python
esports_sorted['AvgEarningsPerYear'] = esports_sorted['TotalMoney'] / esports_sorted['YearsAfterRelease']
esports_sorted['AvgEarningsPerTournament'] = esports_sorted['TotalMoney'] / esports_sorted['TournamentNo']
esports_sorted['AvgEarningsPerPlayer'] = esports_sorted['TotalMoney']/esports_sorted['PlayerNo']
```

```python
esports_sorted.sort_values('AvgEarningsPerYear', ascending=False).head(15)
```

```python
esports_sorted.sort_values('AvgEarningsPerTournament', ascending=False).head(15)
```

```python
esports_total = esports_sorted.groupby('Genre')['TotalMoney'].sum().to_dict()
esports_total = pd.DataFrame.from_dict(esports_total, orient="index")
esports_total.rename(columns={0:"TotalMoney"}, inplace=True)
esports_total = esports_total.sort_values(by="TotalMoney")
esports_total
```

```python
esports_avg_year_total = esports_sorted.groupby('Genre')['AvgEarningsPerYear'].sum().to_dict()
esports_avg_year_total = pd.DataFrame.from_dict(esports_avg_year_total, orient="index")
esports_avg_year_total.rename(columns={0:"AvgEarningsPerYear"}, inplace=True)
esports_avg_year_total = esports_avg_year_total.sort_values(by="AvgEarningsPerYear")
esports_avg_year_total
```

```python
esports_avg_tournament_total = esports_sorted.groupby('Genre')['AvgEarningsPerTournament'].sum().to_dict()
esports_avg_tournament_total = pd.DataFrame.from_dict(esports_avg_tournament_total, orient="index")
esports_avg_tournament_total.rename(columns={0:"AvgEarningsPerTournament"}, inplace=True)
esports_avg_tournament_total = esports_avg_tournament_total.sort_values(by="AvgEarningsPerTournament")
esports_avg_tournament_total
```

```python
esports_avg_player_total = esports_sorted.groupby('Genre')['AvgEarningsPerPlayer'].sum().to_dict()
esports_avg_player_total = pd.DataFrame.from_dict(esports_avg_player_total, orient="index")
esports_avg_player_total.rename(columns={0:"AvgEarningsPerPlayer"}, inplace=True)
esports_avg_player_total = esports_avg_player_total.sort_values(by="AvgEarningsPerPlayer")
esports_avg_player_total
```
